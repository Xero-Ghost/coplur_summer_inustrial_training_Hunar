```python
name = input("Enter student name: ")
cname = input("Enter class: ")

a = float(input("Enter marks of subject 1: "))
b = float(input("Enter marks of subject 2: "))
c = float(input("Enter marks of subject 3: "))
d = float(input("Enter marks of subject 4: "))
e = float(input("Enter marks of subject 5: "))

total = a + b + c + d + e
percentage = total / 5

print("Result")
print("Name = ", name)
print("Class = ", cname)
print("Percentage = ", percentage)
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

str3 = str1 + str2

print("Concatenated string = ", str3)
print("lower() = ", str3.lower())
print("upper() = ", str3.upper())
print("title() = ", str3.title())
print("swapcase() = ", str3.swapcase())
print("capitalize() = ", str3.capitalize())
print("casefold() = ", str3.casefold())
print("center() = ", str3.center(50, "*"))
print("isalnum() = ", str3.isalnum())
print("isdigit() = ", str3.isdigit())
print("isnumeric() = ", str3.isnumeric())
print("isspace() = ", str3.isspace())
print("replace() = ", str3.replace('p', 'x'))
```

```python
a = 10
print("a :", a)
a += 5
print("+= :", a)
a -= 2
print("-= :", a)
a *= 3
print("*= :", a)
a /= 2
print("/= :", a)
a %= 4
print("%= :", a)
a //= 2
print("//= :", a)
a **= 2
print("**= 2 :", a)
```

```python
marks1 = float(input("Enter marks of subject 1: "))
marks2 = float(input("Enter marks of subject 2: "))
marks3 = float(input("Enter marks of subject 3: "))
marks4 = float(input("Enter marks of subject 4: "))
marks5 = float(input("Enter marks of subject 5: "))

total = marks1+marks2+marks3+marks4+marks5
percentage = total/5

if percentage >= 60:
    grade = "A"
elif percentage >= 50:
    grade = "B"
elif percentage >= 40:
    grade = "C"
elif percentage >= 33:
    grade = "D"
else:
    grade = "Fail"

print("Total Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
