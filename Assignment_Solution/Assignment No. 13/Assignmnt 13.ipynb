```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])

arr = np.nan_to_num(arr, nan=0)
print(arr)

print(arr.T)
```

```python
arr = np.arange(24).reshape(2, 3, 4)
print(np.moveaxis(arr, 0, 2))
```

```python
arr = np.array([[1, np.nan, 3],
                [4, 5, np.nan],
                [7, 8, 9]], dtype=float)

col_mean = np.nanmean(arr, axis=0)
idx = np.where(np.isnan(arr))
arr[idx] = np.take(col_mean, idx[1])

print(arr)
```

```python

arr = np.array([-5, 2, -8, 7, -1, 0])
arr = np.where(arr < 0, 0, arr)

print(arr)
```

```python
arr1 = np.array([[3, 4], [5, 6]])
arr2 = np.array([[1, 0], [7, 8]])

avg = (arr1 + arr2) / 2
print(avg)

combined = np.concatenate((arr1.flatten(), arr2.flatten()))

print(np.mean(combined))
print(np.median(combined))

values, counts = np.unique(combined, return_counts=True)
mode = values[np.argmax(counts)]
print(mode)
```

```python
A = np.array([[1, -2, 3],
              [-1, 3, -1],
              [2, -5, 5]])

B = np.array([9, -6, 17])

x = np.linalg.solve(A, B)
print(x)

inv = np.linalg.inv(A)
x2 = np.dot(inv, B)
print(x2)
```

```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Python', 'DBMS', 'DSA', 'English']
sem1 = [78, 85, 74, 80, 88]
sem2 = [82, 79, 81, 84, 90]

x = np.arange(len(subjects))
width = 0.35

plt.bar(x - width/2, sem1, width, label='Semester 1')
plt.bar(x + width/2, sem2, width, label='Semester 2')

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Semester Result Comparison')
plt.xticks(x, subjects)
plt.legend()
plt.show()
```
