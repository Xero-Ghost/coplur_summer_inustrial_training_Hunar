```python
def maximum(a, b, c):
    return max(a, b, c)

print("Maximum:", maximum(10, 25, 15))
```

```python
def distinct_list(lst):
    new_list = []

    for item in lst:
        if item not in new_list:
            new_list.append(item)

    return new_list

print("Distinct List:", distinct_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
def multiply_list(lst):
    result = 1

    for num in lst:
        result *= num

    return result

print("Multiplication:", multiply_list([1, 2, 3, 4]))
```

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact *= i

    return fact

print("Factorial:", factorial(5))
```

```python
def reverse_string(s):
    return s[::-1]

print("Reversed String:", reverse_string("Python"))
```

```python
def check_range(num, start, end):
    if start <= num <= end:
        return "Number is within range"
    else:
        return "Number is outside range"

print(check_range(7, 1, 10))
```

```python
def print_even(lst):
    print("Even Numbers:")

    for num in lst:
        if num % 2 == 0:
            print(num)

print_even([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
def is_prime(n):

    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num = 13

if is_prime(num):
    print(num, "is Prime")
else:
    print(num, "is Not Prime")
```

```python
def count_case(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase Letters:", upper)
    print("Lowercase Letters:", lower)

count_case("Hello Python World")
```

```python
# Writing into a file
file = open("sample.txt", "w")
file.write("This is first line.\n")
file.write("Python File Handling.\n")
file.close()

# Reading from a file
file = open("sample.txt", "r")
content = file.read()
print("File Content:")
print(content)
file.close()

# Appending data into file
file = open("sample.txt", "a")
file.write("This line is appended.")
file.close()

# Reading again after append
file = open("sample.txt", "r")
content = file.read()
print("Updated File Content:")
print(content)
file.close()
```
