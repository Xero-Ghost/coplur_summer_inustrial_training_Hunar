```python
import re

email = "abc@gmail.com"
mobile = "9876543210"
text = "Hello World"
password = "Abc@1234"

email_pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
mobile_pattern = r'^[6-9][0-9]{9}$'
string_pattern = r'^[A-Za-z ]+$'
password_pattern = r'^(?=.*[A-Za-z])(?=.*\d)[A-Za-z\d@$!%*#?&]{8,}$'

print(bool(re.fullmatch(email_pattern, email)))
print(bool(re.fullmatch(mobile_pattern, mobile)))
print(bool(re.fullmatch(string_pattern, text)))
print(bool(re.fullmatch(password_pattern, password)))
```

```python
import pandas as pd

dates = pd.to_datetime(["2026-01-01", "2026-02-15", "2026-03-20"])

df = pd.DataFrame({"date": dates})

print(df["date"].dt.year)
print(df["date"].dt.month)
print(df["date"].dt.day)
print(df["date"].dt.day_name())

sales = pd.DataFrame({
    "date": pd.to_datetime(["2026-01-01", "2026-01-02", "2026-01-03", "2026-01-04"]),
    "sales": [100, 200, 150, 300]
})

sales = sales.set_index("date")
print(sales.resample("D").sum())
print(sales.resample("W").sum())
print(pd.Timedelta(days=5))
```

```python
data = [
    ["2026-01-01", "Laptop", "Electronics", 2, 50000, "Delhi"],
    ["2026-01-02", "Phone", "Electronics", 3, 20000, "Mumbai"],
    ["2026-01-03", "Chair", "Furniture", 5, 3000, "Delhi"],
    ["2026-01-04", "Table", "Furniture", 2, 7000, "Pune"],
    ["2026-01-05", "Laptop", "Electronics", 1, 50000, "Mumbai"],
    ["2026-01-06", "Phone", "Electronics", None, 20000, "Delhi"],
    ["2026-01-07", "Chair", "Furniture", 4, None, "Pune"],
    ["2026-01-08", "Book", "Stationery", 10, 200, "Delhi"],
    ["2026-01-09", "Pen", "Stationery", 20, 10, "Mumbai"],
    ["2026-01-10", "Laptop", "Electronics", 1, 50000, "Delhi"]
]

df = pd.DataFrame(data, columns=["Date", "Product", "Category", "Quantity", "Price", "City"])
df.to_csv("sales_data.csv", index=False)

df = pd.read_csv("sales_data.csv")

print(df.head())

print(df.isnull().sum())

df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
df["Price"] = df["Price"].fillna(df["Price"].median())

df = df.drop_duplicates()
df["Date"] = pd.to_datetime(df["Date"])

df["Revenue"] = df["Quantity"] * df["Price"]

print(df.isnull().sum())

print("Total Revenue:", df["Revenue"].sum())
print("Average Revenue:", df["Revenue"].mean())

print("\nRevenue by Category:")
print(df.groupby("Category")["Revenue"].sum())

print("\nRevenue by City:")
print(df.groupby("City")["Revenue"].sum())

print("\nTop Product:")
print(df.groupby("Product")["Revenue"].sum().sort_values(ascending=False).head(1))

print("\nMonthly Sales:")
print(df.set_index("Date")["Revenue"].resample("ME").sum())
```
