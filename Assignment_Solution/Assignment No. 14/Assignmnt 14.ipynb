```python
def addition(a, b):
    print("Addition Result:", a + b)
```

```python

def subtraction(a, b):
    print("Subtraction Result:", a - b)
def multiplication(a, b):
    print("Multiplication Result:", a * b)

   
```

```python
def division(a,b):
    try:
        if(a>b):
            print("Divsion Result",a/b)
        else:
            print("Divsion Result",b/a)
    except ZeroDivisionError:
        print("the division with zero is not possible")
```

```python
def power(a, b):
    print("Power :", a ** b)

def modulus(a, b):
    print("Modulus Result:", a % b)

def floor_division(a, b):
    try:
        if(a>b):
            print("Floor Division Result:", a // b)
        else:
            print("Floor Division Result:", b // a)
    except ZeroDivisionError:
        print("the division with zero is not possible")
```
