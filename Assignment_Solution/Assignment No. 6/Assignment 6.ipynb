```python
import pandas as pd
data_dict = {
    "A": 10,
    "B": 20,
    "C": 30
}

series_dict = pd.Series(data_dict)

print(series_dict)
print(type(series_dict))

data_list = [100, 200, 300, 400]
series_list = pd.Series(data_list)

print(series_list)
print(type(series_list))



print("Access the elements of a Series in Pandas")
print("First Element:", series_list[0])
print("Second Element:", series_list[1])
print("Last Element:", series_list[3])
```

```python
data_2d = [
    [1, "Aman", 21],
    [2, "Riya", 22],
    [3, "Karan", 20]
]
df1 = pd.DataFrame(data_2d, columns=["ID", "Name", "Age"])
print(df1)

dict_data = {
    "Name": ["Aman", "Riya", "Karan"],
    "Age": [21, 22, 20],
    "City": ["Jaipur", "Delhi", "Mumbai"]
}
df2 = pd.DataFrame(dict_data)
print(df2)

list_of_lists = [
    ["Math", 85],
    ["Science", 90],
    ["English", 88]
]
df3 = pd.DataFrame(list_of_lists, columns=["Subject", "Marks"])
print(df3)

list_of_tuples = [
    ("Aman", 21),
    ("Riya", 22),
    ("Karan", 20)
]

df4 = pd.DataFrame(list_of_tuples, columns=["Name", "Age"])
print(df4)

list_of_dicts = [
    {"Name": "Aman", "Marks": 85},
    {"Name": "Riya", "Marks": 92},
    {"Name": "Karan", "Marks": 78}
]
df5 = pd.DataFrame(list_of_dicts)
print(df5)
```

```python
students = {
    "Name": ["Aman", "Riya", "Karan"],
    "Age": [21, 22, 20],
    "Marks": [85, 92, 78]
}
df = pd.DataFrame(students)

print("Using iterrows()")
for index, row in df.iterrows():
    print(index, row["Name"], row["Marks"])
print("Using itertuples()")
for row in df.itertuples():
    print(row.Name, row.Age, row.Marks)

result = df[df["Marks"] > 80]
print("Marks > 80",result)

print(df.iloc[1])

print(df.loc[0:1, ["Name", "Marks"]])

new_df = df[df["Marks"] >= 80]
print("Dataframe on conditions",new_df)

new_row = pd.DataFrame({
    "Name": ["Neha"],
    "Age": [23],
    "Marks": [95]
})
print("before",new_row)
top = df.iloc[:1]
bottom = df.iloc[1:]
df_updated = pd.concat([top, new_row, bottom]).reset_index(drop=True)
print("after",df_updated)

row_list = df.values.tolist()
print(row_list)
print(type(row_list))
```
